```python
import scipy.io.wavfile as wavfile
import scipy
import scipy.fftpack as fftpk
import numpy as np
from matplotlib import pyplot as plt

s_rate, signal = wavfile.read("nada.wav")

FFT = abs(scipy.fft(signal))
freqs = fftpk.fftfreq(len(FFT), (1.0/s_rate))
plt.plot(freqs[range(len(FFT)//2)], FFT[range(len(FFT)//2)])
plt.xlabel('Time (s)')
plt.ylabel('Magnitude')
```

```python
plt.plot(FFT)
```

```python
irfft_output = np.fft.ifft(FFT)
```

```python
irfft_output = np.real(irfft_output)
```

```python
plt.plot(irfft_output)
```

```python
scale = np.int16(irfft_output/np.absolute(irfft_output).max()*50000)
```

```python
wavfile.write('inversed.wav',s_rate,scale)
```
